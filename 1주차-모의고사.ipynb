### 모의고사 1회 문제 - 1유형
#### 1. 

```python
import pandas as pd
import numpy as np

df = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/dataset/main/airquality.csv')
df
```

```python
df.isnull().sum() # 결측치가 37개로 가장많다. 
df.info()
# 총 153개 중 Ozone은 116개가 non-null로 none이 가장많다. 
# 따라서 Ozone의 mean 구하고 none을 0으로 대치하고 mean을 구함.

df["Ozone0"] = df["Ozone"].fillna(0, inplace = False)
print(df["Ozone"].mean() - df["Ozone0"].mean())
# 답: 10.18813
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

scaler1 = StandardScaler()
scaler2 = MinMaxScaler()
scaler2.fit_transform(df[["Wind"]]).mean() - scaler1.fit_transform(df[["Wind"]]).mean()
# 답: 0.434606
```

```python
import numpy as np
import pandas as pd
import sklearn

df = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/dataset/main/HR-Employee-Attrition.csv')
df
```

```python
df['Attrition_scalar'] = df['Attrition'].replace({'Yes':1, 'No':2}, inplace = False)
df.groupby('Attrition_scalar').size()
# 1: 237개, 2: 1233개
```

#### 2. 

```python
df.info()
# : int64(27), object(9) 개 
cat_feat = df.select_dtypes('object','category').columns.values
df[cat_feat].nunique().sort_values()
# Over18  : 1개
df.drop('Over18', axis = 1)
```

#### 3. 

```python
df = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/dataset/main/HR-Employee-Attrition.csv')
df_num = df.select_dtypes('number') #숫자형 데이터 뽑기
corr_mat = df_num.corr(method = 'pearson')#person corr계산
corr_mat.shape # 26x26임을 확인 
col = list(corr_mat.columns)
for i in range(26):
    for j in range(26):
        if i < j and corr_mat.iloc[i, j] >= 0.9:
            print(col[i], col[j])

# JobLevel MonthlyIncome 
df_num.drop('JobLevel', axis = 1)
```
